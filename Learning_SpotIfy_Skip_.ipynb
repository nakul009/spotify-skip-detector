```python
import pandas as pd
csv_file = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-01-21/spotify_songs.csv'

print ('Fetching data from github CSV file.. ')

df = pd.read_csv(csv_file)

print("successfully read the file")

print ("Data preview")

display(df.head())
```

```python
df['playlist_genre'].nunique()
```

```python
print("Transalating playlist genre into binary checkboxes as column ....")

df_encoded = pd.get_dummies(df, columns=['playlist_genre'])

print("Translating Complete!")

display(df_encoded.head())
```

```python
df_encoded['track_name'].nunique()
```

```python
df_encoded['playlist_subgenre'].nunique()
```

```python
# --- PHASE 2A: THE REPUTATION SCORE (Target Encoding) ---
print("Calculating artist reputation scores...\n")

df_encoded['track_artist'] = df_encoded['track_artist'].fillna('Unknown')

# 1. Group by artist and calculate their average 'track_popularity'
artist_means = df_encoded.groupby('track_artist')['track_popularity'].mean()

# 2. Map those average scores back to our dataset as a brand new column
df_encoded['artist_popularity_score'] = df_encoded['track_artist'].map(artist_means)

print("Target Encoding Complete! Look at the Artist Name vs their new Reputation Score:")
display(df_encoded[['track_artist', 'artist_popularity_score']].head(10))

''''
Ed sheeran
 -- Song 1 100
 -- Song 2 50
  -- Song 3 50

  Create a mean value from all the songs from Ed sheeran the result to the track artist column
'''
```

```python
display(df_encoded.head())
```

```python
# --- PHASE 2B: PREVENTING CHEATING (Creating X and y) ---
print("Preventing Data Leakage: Separating questions from the answer key...\n")

# 1. 'y' is our Answer Key
# We define a 'Listen' (1) as a popularity over 50, and a 'Skip' (0) as under 50.
y = (df_encoded['track_popularity'] > 50).astype(int)

# 2. 'X' is our Test Questions (Everything EXCEPT the answer key)
# We MUST drop the exact popularity score here so the AI doesn't cheat!
X = df_encoded.drop(columns=['track_popularity'])

print("Answer Key (y) and Test Questions (X) separated successfully!")
print(f"We have {X.shape[1]} test questions ready to go.")

'''
80% of our data

and 20% is for testing our model

X. - Quetions or our test data

y - Answers populatiry score
'''
```

```python
columns_to_drop = [
    'track_id',
    'track_name',
    'track_artist',
    'track_album_id',
    'track_album_name',
    'track_album_release_date',
    'playlist_name',
    'playlist_id',
    'playlist_subgenre'    # Removing this to avoid the "Math Trap"
]

# Tell pandas to drop them from our Test Questions (X)
X_final_data_only = X.drop(columns=columns_to_drop)

print("Here is our clean, math-only dataset:")
display(X_final_data_only.head())
```

```python
print(X_final_data_only.isnull().sum())
```

```python
from sklearn.model_selection import train_test_split

# --- PHASE 3A: THE 80/20 SPLIT ---
print("Slicing the data into 80% Training and 20% Testing...\n")

# This takes our clean math questions (X_final_data_only) and answers (y) and shuffles them into piles
X_train, X_test, y_train, y_test = train_test_split(X_final_data_only, y, test_size=0.2, random_state=42)

print(f"Training pile (X_train): {X_train.shape[0]} songs")
print(f"Testing pile (X_test): {X_test.shape[0]} songs")
```

```python
from sklearn.preprocessing import MinMaxScaler

# --- PHASE 3B: THE EQUALIZER (Feature Scaling) ---
print("Scaling features so all numbers are on a level playing field (0 to 1)...\n")

# 1. Initialize the MinMaxScaler tool
scaler = MinMaxScaler()

# 2. Fit and transform the training data
# (This measures the min/max of X_train AND scales it all at once)
X_train_scaled = scaler.fit_transform(X_train)

# 3. ONLY transform the test data
# (This scales X_test using the exact measurements we learned from X_train)
X_test_scaled = scaler.transform(X_test)

print("Scaling complete! Let's look at a sample of our perfectly balanced data:")
print(X_train_scaled[0])
```

```python
import pandas as pd

# Convert the scaled array back into a DataFrame so we can see column names
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=X_train.columns)

display(X_train_scaled_df.head())
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout

model = Sequential()

# Layer 1 first later
model.add(Dense(units = 32, activation='relu', input_dim=(X_train_scaled_df.shape[1])))

# Layer 2
model.add(Dense(units = 16, activation='relu'))

# Layer 3

model.add(Dense(units = 1, activation='sigmoid'))

model.summary()
"""
output -8 ---> 0.00003

output 4 --> 0.92428472 92.42%
"""
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
"""
prediction - 0.95 (95% probability user will listen to the song).  correct answre = 0 (Skip the song)
loss is used by neural netrowrk to lean

accuracy is used by us to monitor how well the model is doing
"""
```

```python
model.fit(X_train_scaled_df, y_train, epochs=20, batch_size=32)
```
